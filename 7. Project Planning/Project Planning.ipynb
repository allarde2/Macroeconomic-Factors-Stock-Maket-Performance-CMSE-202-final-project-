## PROJECT PLANNING

````python
# Update the Gantt chart to reflect the more model-focused and clearly scoped project

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

updated_tasks = {
    "Define scope & research questions": ("2025-03-25", "2025-03-27"),
    "Find & download quarterly datasets (macro + market)": ("2025-03-25", "2025-03-29"),
    "Set up GitHub repo & project structure": ("2025-03-25", "2025-03-26"),
    "Clean & merge datasets": ("2025-03-29", "2025-04-03"),
    "Exploratory data analysis (EDA)": ("2025-04-01", "2025-04-05"),
    "Build ARIMA model": ("2025-04-04", "2025-04-08"),
    "Build Multiple Linear Regression model": ("2025-04-06", "2025-04-09"),
    "Build Lasso regression models": ("2025-04-09", "2025-04-13"),
    "Model comparison & evaluation": ("2025-04-13", "2025-04-16"),
    "Finalize visualizations": ("2025-04-15", "2025-04-18"),
    "Build slides & draft narrative": ("2025-04-17", "2025-04-20"),
    "Rehearsals & polish": ("2025-04-20", "2025-04-21"),
}

updated_df = pd.DataFrame([
    {"Task": task, "Start": pd.to_datetime(start), "End": pd.to_datetime(end)}
    for task, (start, end) in updated_tasks.items()
])

# Plot updated Gantt chart
fig, ax = plt.subplots(figsize=(11, 6))
for i, row in updated_df.iterrows():
    ax.barh(row["Task"], (row["End"] - row["Start"]).days, left=row["Start"], color="lightgreen")

ax.set_xlabel("Date")
ax.set_title("Updated Project Gantt Chart")
ax.xaxis.set_major_locator(mdates.WeekdayLocator(byweekday=mdates.MO))
ax.xaxis.set_major_formatter(mdates.DateFormatter("%b %d"))
plt.tight_layout()
plt.gca().invert_yaxis()
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
````
