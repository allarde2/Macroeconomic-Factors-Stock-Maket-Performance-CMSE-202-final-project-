<p style="font-family: 'Times New Roman'; font-size: 18px; text-align: center; color: black;">
  For our project, the goal of using Multiple Linear Regression is to predict the value of our dependent variable (S&P 500/Dow Jones/NASDAQ) based on our independent variables (CPI, Unemployment Rate, Gross Domestic Product Growth, M2 Money Supply, and Federal Funds Rate). To begin, we will look over the mathematical background of Multiple Linear Regression.
</p>

<div style="text-align: center;">
  
</div>

<p style="font-family: 'Times New Roman'; font-size: 18px; text-align: center; color: black;">
  Multiple Linear Regression uses Ordinary Least Squares (OLS) to minimize the <strong>loss function</strong>, which reduces the sum of the squared residuals. OLS helps estimate the y value predictions by choosing coefficient values that make these values as close as possible to the actual values. These coefficients values tell us the type of relationship (positive or negative) and the strength (weak or strong) of the relationship between each independent variable and the dependent variable.
</p>

<div style="text-align: center;">
  
</div>

<p style="font-family: 'Times New Roman'; font-size: 18px; text-align: center; color: black;">
  For Multiple Linear Regression, we will use the <strong>statsmodel</strong> package and the <strong>.fit()</strong> argument to compute the OLS model. This is solving for the best beta values (coefficient values) that will minimize the loss function. The <strong>.predict()</strong> argument will make the predictions using the y equals function above.
</p>
