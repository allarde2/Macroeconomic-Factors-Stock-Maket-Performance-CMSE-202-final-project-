<p style="font-family: 'Times New Roman'; font-size: 18px; text-align: center; color: black;">
  We will proceed running a Lasso Regression analysis on the differenced dependent variable Dow Jones. We will begin by setting the libraries that we will use in this analysis:
</p>

<div style="font-family: 'Times New Roman'; font-size: 18px; text-align: center; color: black;">
  <ul style="display: inline-block; text-align: left;">
    <li><strong>import pandas as pd</strong>: loading in data, working with pandas data frames, and differencing the data</li>
    <li><strong>import numpy as np</strong>: Built in numpy functions</li>
    <li><strong>from sklearn.model_selection import train_test_split</strong>: split our data into test/train using shuffle = False due to time-series data</li>
    <li><strong>from sklearn.metrics import mean_squared_error</strong>: calculate MSE and RMSE scores</li>
    <li><strong>from sklearn.linear_model import LassoCV</strong>: run the LASSO regression</li>
    <li><strong>from sklearn.preprocessing import StandardScaler</strong>: Scaling the data for LASSO</li>
    <li><strong>import matoplotlib.pyplot as plt</strong>: Plotting Actual versus Prdicted Test Data</li>
  </ul>
</div>

```python
# Importing libraries for time series modeling and evaluation 

import pandas as pd  # Data manipulation and analysis
import numpy as np  # Numerical operations and array handling
from sklearn.model_selection import train_test_split  # Splitting data into training and testing sets
from sklearn.metrics import mean_squared_error  # Model evaluation using MSE/RMSE
from sklearn.linear_model import LassoCV  # Lasso regression with cross-validation for regularization
from sklearn.preprocessing import StandardScaler  # Standardizing features
import matplotlib.pyplot as plt  # Plotting and visualization
```

<p style="font-family: 'Times New Roman'; font-size: 18px; text-align: center; color: black;">
  Then, we will load in our data and rename column names for easier and more descriptive analysis.
</p>

```python
# Load the merged dataset from a CSV file
data = pd.read_csv('merged_data.csv')

# Rename columns for clarity and consistency across analysis
data.rename(columns={
    'CPALWE01USQ661N': 'CPI',
    'S&P 500 ^GSPC': 'S&P_500',
    'NASDAQ ^IXIC': 'NASDAQ',
    'Dow Jones ^DJI': 'Dow_Jones',
    'RealInterestRate': 'Real_IR',
    'UnemploymentRate': 'Unemployment_Rate',
    'FedFundsRate': 'Federal_Funds_Rate'
}, inplace = True)

# Saves the 'Date' column separately for future plotting 
dates = data['Date']

# Displays the full DataFrame 
data
```

<p style="font-family: 'Times New Roman'; font-size: 18px; text-align: center; color: black;">
  Then, as explained in the Time-series Backrgound notebook we will be differencing the data using the built in python function 
    <strong>.diff()</strong>. Due to differencing, the first row will be NA; therefore, we will drop the corresponding row. 
</p>

```python
# Apply first-order differencing to the Dow Jones index to remove trend and stationarize the series
# This follows an ARIMA(0,1,0) structure: no autoregressive or moving average terms, just differencing
data['Dow_Jones_Diff'] = data['Dow_Jones'].diff()

# Drops the first row with NaN resulting from the differencing operation
data = data.dropna()

# Displays the updated dataset with differenced Dow Jones values
data
```

<p style="font-family: 'Times New Roman'; font-size: 18px; text-align: center; color: black;">
  Then, we will begin running our Lasso Regressive Model using the differencing data. For our independent variables, it will include all macroeconomic variables and the Dow Jones differencing column and we will select all rows from April 1 1992 - July 1 2023. This represents t-1, the previous quarter, for our time-series analysis. For our dependent variable, we will be selecting the Dow Jones differencing column from July 1 1992 - October 1 2023. This represents t, the current quarter, for our time-series analysis. 
</p>

```python
# Define independent and dependent variables for time series prediction
# We use lagged macroeconomic indicators and Dow_Jones_Diff at time t-1 to predict Dow_Jones_Diff at time t

independent_variables = ['CPI', 'Real_IR', 'Unemployment_Rate', 'GDP_Growth', 
                         'M2_Supply', 'Federal_Funds_Rate', 'Dow_Jones_Diff']
dependent_variable = ['Dow_Jones_Diff']

# Sets up predictors (t-1 values)
x = data[independent_variables].iloc[0:-1]  # Excludes last row

# Sets up response variable (t values)
y = data[dependent_variable].iloc[1:]  # Excludes first row
```

<p style="font-family: 'Times New Roman'; font-size: 18px; text-align: center; color: black;">
  Then, using <strong>sklearn</strong>, we split our data into an 80/20 train-test split with <strong>shuffle = False</strong>, as done in previous notebooks.
  We ran LASSO regression similarly to the Dow Jones notebook, where we transform <code>x_train</code> and <code>x_test</code> to be <strong>standard normal</strong>, meaning the mean equals 0 and the standard deviation equals 1.
</p>

<p style="font-family: 'Times New Roman'; font-size: 18px; text-align: center; color: black;">
  For <strong>LassoCV</strong>, we set <strong>cv = 5</strong> and <strong>random_state = 42</strong> to ensure consistent results.
  In our analysis, we focused on:
</p>

<ul style="font-family: 'Times New Roman'; font-size: 18px; color: black; text-align: center; list-style-type: none; padding-left: 0;">
  <li>→ <strong>RMSE</strong> score to compare model accuracy</li>
  <li>→ Optimal <strong>lambda</strong> to see how quickly variables approach zero</li>
  <li>→ Final <strong>beta values</strong> to interpret strength and direction of predictors</li>
</ul>

```python
# Split the time series data into training and testing sets (80/20 split),
# maintaining the chronological order by setting shuffle = False
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, shuffle = False)

# Adjusts the corresponding dates for the test set to use in future plotting
dates = dates.iloc[y_test.index]
```

```python
# Standardize the independent variables: transform to standard normal (mean = 0, std = 1)
# This step is important for Lasso regression, which is sensitive to feature scale
scaler = StandardScaler()

# Fits the scaler on the training data and apply the transformation
x_train_scaled = scaler.fit_transform(x_train)

# Applies the same transformation to the test data using the fitted scaler
x_test_scaled = scaler.transform(x_test)
```

```python
# Run Lasso regression with 5-fold cross-validation on the training set
lasso = LassoCV(cv = 5, random_state = 42)

# Fit the model using the standardized training data
# Used .ravel() to flatten y and suppress shape warnings (Consulted Generative AI)
lasso.fit(x_train_scaled, y_train.to_numpy().ravel())

# Retrieves the optimal regularization parameter (alpha) selected via cross-validation
optimal_lambda = lasso.alpha_

# Extracts the model coefficients for interpretation
coefficient = lasso.coef_
```

```python
# Generate predictions on the test set using the trained Lasso model
y_pred = lasso.predict(x_test_scaled)

# Calculates Root Mean Squared Error (RMSE) to evaluate prediction accuracy
rmse = (mean_squared_error(y_test, y_pred)) ** 0.5

# Calculates Mean Squared Error (MSE) as the square of RMSE
mse = rmse ** 2
```

```python
# Extracts individual Lasso coefficients for interpretation
cpi = coefficient[0]
real_ir = coefficient[1]
unemployment = coefficient[2]
gdp_growth = coefficient[3]
m2_supply = coefficient[4]
fed_funds = coefficient[5]
dow_diff = coefficient[6]

# Prints model performance and coefficients in a formatted summary
print(f'Test RMSE:{rmse:.2f}---Optimal Lambda:{optimal_lambda:.2f}---'
      f'CPI:{cpi:.2f}---Real Interest Rate:{real_ir:.2f}---Unemployment Rate:{unemployment:.2f}---'
      f'GDP Growth:{gdp_growth:.2f}---M2 Money Supply:{m2_supply:.2f}---Federal Funds Rate:{fed_funds:.2f}---'
      f'Dow Jones Difference:{dow_diff:.2f}-')
```

<p style="font-family: 'Times New Roman'; font-size: 18px; text-align: center; color: black;">
  Using our coefficients generated from lasso, the three most important predictors for Dow Jones using differenced data was Unemployment Rate, GDP Growth, and M2 Money Supply. This is not far off from what we found out in Multiple Linear Regression and Lasso Regressive analysis. However, our new predictor for time-series, Dow Jones Difference was not a positive driver, or any driver at all for Dow Jones, meaning that predicting the current quarter using the past quarter is not helpful. To visualize, we will plot Actual vs Predicted Dow Jones for time-series, which will look different compared to our Multiple Linear Regression Plot because of the differencing componenent. This makes Dow Jones more sporadic and less smooth, showing the raw changes in the index.
</p>

```python
# Plot actual vs predicted differenced Dow Jones values using the time-series Lasso model
# (Plot aesthetic was consulted via generative AI)

plt.figure(figsize = (12, 7))  # Set the figure size for clarity

# Plot actual values from the test set
plt.plot(dates, y_test, 
         label = 'Actual', 
         color = 'royalblue', 
         linewidth = 2.5)

# Plot predicted values from the Lasso model
plt.plot(dates, y_pred, 
         label = 'Predicted (Time-Series)', 
         color = 'deepskyblue', 
         linewidth = 2.5, 
         linestyle = '--')

# Add title and axis labels with custom styling
plt.title('Actual vs Predicted Dow Jones (Time-Series)', 
          fontsize = 18, 
          fontweight = 'bold', 
          pad = 15, 
          color = 'midnightblue')

plt.xlabel('Date', fontsize = 14, color = 'navy')
plt.ylabel('S&P 500', fontsize = 14, color = 'navy')

# Format tick labels
plt.xticks(rotation = 45, fontsize = 12, color = 'navy')
plt.yticks(fontsize = 12, color = 'navy')

# Set plot background color
ax = plt.gca()
ax.set_facecolor('white') 
ax.figure.set_facecolor('white') 

# Customize plot borders (spines)
for spine in plt.gca().spines.values():
    spine.set_visible(True)
    spine.set_linewidth(1.25)
    spine.set_color('black')

plt.xlabel('Date', fontsize = 14, color = 'navy')
plt.ylabel('S&P 500', fontsize = 14, color = 'navy')

# Format tick labels
plt.xticks(rotation = 45, fontsize = 12, color = 'navy')
plt.yticks(fontsize = 12, color = 'navy')

# Set plot background color
ax = plt.gca()
ax.set_facecolor('white') 
ax.figure.set_facecolor('white') 

# Customize plot borders (spines)
for spine in plt.gca().spines.values():
    spine.set_visible(True)
    spine.set_linewidth(1.25)
    spine.set_color('black')

# Add gridlines to enhance readability
plt.grid(visible = True, linestyle = '--', linewidth = 0.5, alpha = 0.7, color = 'lightblue')

# Add legend with styling
plt.legend(loc = 'upper left', fontsize = 12, frameon = True, shadow = True, edgecolor = 'steelblue')

# Adjust layout to avoid label cutoff
plt.tight_layout()

# Displays the plot
plt.show()
```

This may be due to the fact that the independent variables were differenced, and macroeconomic indicators typically do not change as rapidly as stock market values.
